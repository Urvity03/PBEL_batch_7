```python
# conditional statement

age = 18

if age >= 18:
    print("you are eligible for voting")

elif age == 18:
    print("you are eligible for voting")

else:
    print("you are not eligible for voting")
```

```python
#looping statement for loop and while loop
list1 = [1,2,3,4,5]
for i in list1:
    print(i)
for i in range(4):
    print(i)
    
```

```python
#while loop
# print numbers from 0 to 3

i = 0

while i < 4:
    print(i)
    i += 1
```

```python
#create a program where user enter a number and program will
# add all number one by one and print the result until
# the user will enter q ro quit the program
#hint while loop and break statement, start with sum=0 and add number to the sum
```

```python
# add numbers until user enters q

total = 0

while True:
    num = input("Enter a number (or q to quit): ")

    if num.lower() == "q":
        break

    total += int(num)

print("Sum of all numbers entered:", total)
```
